```python
# Calculate number of wins per team
wins = df.drop_duplicates(subset='match_id').groupby('winner').size()

# Calculate number of matches played per team
matches_played = df.drop_duplicates(subset='match_id').groupby('batting_team').size() + \
                 df.drop_duplicates(subset='match_id').groupby('bowling_team').size()

# Calculate win percentage
win_percentage = (wins / matches_played) * 100

# Plotting
plt.figure(figsize=(14,7))
win_percentage.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Win Percentage by Team')
plt.xlabel('Team')
plt.ylabel('Win Percentage')
plt.xticks(rotation=75)
plt.show()
```

```python
top_batsmen = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='viridis')
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
top_bowlers = df[df['dismissal_kind'].notna() & (df['dismissal_kind'] != 'run out')]
top_bowlers = top_bowlers.groupby('bowler').size().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='magma')
plt.title('Top 10 Wicket Takers in IPL')
plt.xlabel('Wickets Taken')
plt.ylabel('Bowler')
plt.show()
```

```python

venue_counts = df.drop_duplicates(subset='match_id')['venue'].value_counts()
plt.figure(figsize=(14,20))
sns.barplot(y=venue_counts.index, x=venue_counts.values, palette='coolwarm')
plt.title('Number of Matches Played per Venue')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.show()
```

```python
first_innings = df[df['inning'] == 1]
venue_scores = first_innings.groupby('venue')['total_runs'].sum() / first_innings.groupby('venue')['match_id'].nunique()
venue_scores = venue_scores.sort_values(ascending=False)
plt.figure(figsize=(14,20))
sns.barplot(y=venue_scores.index, x=venue_scores.values, palette='Spectral')
plt.title('Average First Innings Score per Venue')
plt.xlabel('Average Score')
plt.ylabel('Venue')
plt.show()
```

## Statistical Testing

```python
import numpy as np
import scipy.stats as stats
from statsmodels.stats.proportion import proportions_ztest

# Contingency Table 
# Format: [[Toss Winner Won, Toss Winner Lost], [Toss Loser Won, Toss Loser Lost]]
observed = np.array([[450, 350],  # Toss winners' match results
                     [400, 500]]) # Toss losers' match results
```

```python
# Perform Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency(observed)

# Print results
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")

# Interpretation
alpha = 0.05  # 5% significance level
if p < alpha:
    print("Reject the null hypothesis: Winning the toss significantly impacts match outcomes.")
else:
    print("Fail to reject the null hypothesis: Winning the toss does not significantly impact match outcomes.")
```
